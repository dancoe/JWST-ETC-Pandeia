# Calculations

```python
# Start at bright magnitude that saturates grism at all wavelengths
# Gradually make fainter and measure when unsaturated: that's the limit
# May take a while depending on how small you make each dmag step

nodata = -99

#for key in 'g2v m2v a0v'.split():
#for star_type in ['a0v']:
for star_type in ['m2v']:
    print('=============')
    print(star_type)
    calculation['scene'][0]['spectrum']['sed']['key'] = star_type

    for filt in 'f322w2 f444w'.split():
        outfile = 'grismsaturation_%s_%s.txt' % (filt, star_type)
        if os.path.exists(outfile):
            print(outfile, 'EXISTS')
            continue
        
        print('=============')
        print(filt)

        if filt == 'f322w2':
            mag = 3  # F322W2
        else:
            mag = 2  # F444W

        calculation['configuration']['instrument']['filter'] = filt
        calculation['scene'][0]['spectrum']['normalization']['norm_flux'] = mag
        results = perform_calculation(calculation)
        slam, sat = results['1d']['n_full_saturated']
        
        limits = slam * 0 + nodata
        ii = np.arange(len(limits))

        dmag = 0.01
        while (limits < -90).any():
            calculation['scene'][0]['spectrum']['normalization']['norm_flux'] = mag
            results = perform_calculation(calculation)
            slam, sat = results['1d']['n_full_saturated']
            slam, sn  = results['1d']['sn']

            print()
            print('mag = %.2f' % mag)
            #print('sat, snr:')
            for i, lam in enumerate(slam):
                #print('%.1f'%lam, sat[i], sn[i])
                if sat[i] == 0:  # not saturated
                    if limits[i] == nodata:
                        limits[i] = mag

            print('limits:')
            #for i in ii[-20:]:
            for i in ii[::100]:
                print('%.4f'%slam[i], '%.2f'%limits[i])

            mag += dmag
            
        #filt = calculation['configuration']['instrument']['filter']
        #stype = calculation['scene'][0]['spectrum']['sed']['key']

        print('SAVING', outfile)
        ascii.write([slam, limits], outfile, names='lambda mag'.split())

        fig = plt.figure(figsize=(8,6))
        plt.plot(slam, limits)
        plt.ylim(plt.ylim()[::-1])

        plt.xlabel('Wavelength (microns)')
        plt.ylabel('Saturation K mag (Vega) for %s star' % key.upper())
        outfile = outfile.replace('.txt', '.png')
        plt.savefig(outfile)
```
